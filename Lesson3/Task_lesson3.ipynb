3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
1. окружности,
2. эллипса,
3. гиперболы.

```python
#элипс
b=2
a=1
x=[]
y1=[]
y2=[]
for i in range (-1000, 1001):
    x_=i/1000
    x.append (x_)
    y1.append (b*math.sqrt(1-x_**2/a**2))
    y2.append (-b*math.sqrt(1-x_**2/a**2))
    
rcParams["figure.figsize"] = 9, 4
plt.plot (x,y1)
plt.plot (x,y2)
plt.xlabel ('x')
plt.ylabel ('y')
plt.show()
```

```python
#окружность
r=1
x_0 =2
y_0=3
x=[]
y1=[]
y2=[]
for i in range (-100, 101):
    x_=i/100
    x.append (x_)
    y1.append (y_0+math.sqrt(r**2-(x_)**2))
    y2.append (y_0-math.sqrt(r**2-(x_)**2))

rcParams["figure.figsize"] = 9, 9
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
#гипербола
b=2
a=1
x=[]
y1=[]
y2=[]
for i in range (-1000*a-10000, 1000*a+10000):
    if i>1000*a or i<-1000*a:
        x_=i/1000
        x.append (x_)
        y1.append (b*math.sqrt(x_**2/a**2-1))
        y2.append (-b*math.sqrt(x_**2/a**2-1))
    
rcParams["figure.figsize"] = 9, 9
plt.plot (x,y1)
plt.plot (x,y2)
plt.xlabel ('x')
plt.ylabel ('y')
plt.show()
```

1) Решите систему уравнений:
y = x 2 – 1

exp(x) + x∙(1 – y) = 1

```python
x = np.linspace(-2, 3, 201)
plt.plot(x, (np.exp(x) +x- 1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-10,10) 
plt.grid(True)
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x*(1-y)-1)

x1, y1 =  fsolve(equations, (-2, 1))
              
print (x1, y1)
x2, y2 =  fsolve(equations, (2, 4))
              
print (x2, y2)
```
